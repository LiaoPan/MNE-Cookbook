#### 删除Raw数据（脑磁数据）的部分通道

```python
import mne
print("MNE VERSION:",mne.__version__)
raw = mne.io.read_raw_fif("/Users/reallo/Downloads/anonymize_raw_tsss.fif",verbose=0)
```

```python
raw.plot();
```

```python
raw.info['ch_names']
```

```python
# https://mne.tools/stable/generated/mne.io.Raw.html
raw.drop_channels(['ECG061']) # 删除ECG061通道
```

```python
raw.save("no_ecg_raw.fif")
```

```python
raw.drop_channels(['EOG062']) # 删除EOG062通道
```
